```python
import numpy as np, random, operator, pandas as pd, string
import matplotlib.pyplot as plt
```

### Initialise population

```python
# Create the class DNA
class DNA:
    
    def __init__(self):
        
        # Create a random string of lowercase characters length = 18 
        self.length = 18
        
        # Crete a floating point fitness variable
        self.fitness = 0.0
        
        # Initilaise genes to be a string and then populate with
        # random characters between ascii range 32 -> 128
        self.genes = ''
        for ii in range(self.length):
            self.genes += chr(random.randrange(32, 128))
        
    def calculateFitness(self):
        
        # Score variable to count how many matches there are
        score = 0
        
        # Check matches between current phrase and target phrase
        for ii in range(self.length):
            if self.genes[ii] == target[ii]:
                score += 1
        
        # Fitness is number of mathches / length of thread
        self.fitness = float(score)/len(target)
        
        
    def crossover(self, partner):
        
        # Create the child and then overwrite it later on
        child = DNA()
        
        # Pick a midpoint randomly 
        midpoint = random.randrange(self.length)
        
        # Temporarily store the string and then update child.genes after
        tempString = ''
        
        # Before the midpoint use parent1, after use parent2
        for ii in range(self.length):
            if ii > midpoint: 
                tempString += self.genes[ii]
            else:
                tempString += partner.genes[ii]
        
        # Update genes
        child.genes = tempString
        
        return child
    
    def mutate(self, mutationRate):
        
        # For every character there is a chance of mutation
        for ii in range(self.length):
            if random.uniform(0, 1) < mutationRate:
                
                # Use a list to edit the string
                stringList = list(self.genes)
                stringList[ii] = chr(random.randrange(32, 128))
                self.genes = ''.join(stringList)
```

Now create a setup function to create individual DNA's and add them to the population array.

```python
def setup(totalPopulation):
    # Loop through and add a new DNA instance to the population
    for ii in range(totalPopulation):
        population.append(DNA())
```

Let's create a draw function which calculates the fitness of all the DNA in population.

```python
def draw():
    # Calculate fitness for every DNA in population
    for ii in range(len(population)):
        population[ii].calculateFitness()
```

Create the mating pool.

```python
def populateMatingPool():
    # Add each member n times according to it's fitness score
    for ii in range(len(population)):
        n = int(population[ii].fitness * 1000);
        for jj in range(n):
            matingPool.append(population[ii])

    # Check if mating pool is empty and if so then evenly distribute
    # mating pool.
    if len(matingPool) == 0:
        for jj in range(len(population)):
            matingPool.append(population[ii])
```

### Now put it all together

```python
# Set random seed
random.seed(10)

# Initialise variables that can be changed
totalPopulation = 150
mutationRate = 0.01
target = "to be or not to be"

# Initialise other variables
population = []
matingPool = []
bestPhrase = ''
currentGeneration = 0

# Setup population array and calculate fitness for every DNA
setup(totalPopulation)
draw()

# Loop until you have the phrase
while bestPhrase != target:

    # Increment currentGeneration
    currentGeneration += 1

    # Highest fitness variable to keep track of bestPhrase
    # in this current generation
    highestFitness = 0
    totalFitness = 0.0

    # Calculate fitness for every dna and populate mating pool
    # based on fitness
    draw()

    # Empty mating pool and then re populate it based on fitness
    matingPool = []
    populateMatingPool()    

    for ii in range(len(population)):

        # Calculate totalFitness for averageFitness
        totalFitness += population[ii].fitness

        # Choose 2 random parents from the mating pool
        parentA = random.choice(matingPool)
        parentB = random.choice(matingPool)

        # Crossover and then mutate and then calculate fitness
        child = parentA.crossover(parentB)
        child.mutate(mutationRate)
        child.calculateFitness()

        # Check if child has the highest fitness and update
        # bestPhrase if so
        if child.fitness > highestFitness:
            bestPhrase = child.genes
            highestFitness = child.fitness

        # Update population array 
        population[ii] = child    

    # Calculate average fitness for this generation
    averageFitness = totalFitness / float(len(population))

    # Print out statistics from each generation
    print ("Current generation = " + str(currentGeneration))
    print ("Average fitness = " + str(averageFitness))
    print ("Best phrase = " + bestPhrase)
        
```
